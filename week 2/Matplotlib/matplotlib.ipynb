# Stack Plot

```python
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
playing=[8,5,7,8,13]
working=[7,8,7,2,2]
plt.plot([],[],color='m',label='Sleeping',linewidth=5)
plt.plot([],[],color='m',label='Eating',linewidth=5)
plt.plot([],[],color='m',label='Working',linewidth=5)
plt.plot([],[],color='m',label='playing',linewidth=5)
plt.stackplot(days,sleeping,eating,working,playing)
colors=(['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

# Pie chart

```python
slices=[7,2,2,13]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','b']

plt.pie(slices,
       labels=activities,
       colors=cols,
       startangle=90,
       shadow=True,
       explode=(0,0.1,0,0),
       autopct='%1.1f%%')
plt.title('Pie Plot')
plt.show()
```

# Working with multiple plots

```python
import numpy as np
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.subplot(211)     #helps to plot multiple plots horizonatally 2 plots horizontally 1 plots
plt.plot(t1,f(t1),'bo',t2,f(t2))

plt.subplot(212)   # 2 graphs,align vertical or horizontal, which graph presesnt
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
